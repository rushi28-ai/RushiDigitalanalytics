#### Partitioning Example 1

```sql
DROP TABLE IF EXISTS users;

CREATE TABLE users (
  id INT,
  name STRING,
  age INT,
  gender STRING
)
```

```sql
INSERT INTO users VALUES (1, "Raju", 48, "Male"), (2, "Ramesh", 25, "Male"),(3, "Ramya", 38, "Female");
INSERT INTO users VALUES (4, "Ravi", 28, "Male"), (5, "Raheem", 25, "Male"), (6, "Radha", 45, "Female");
INSERT INTO users VALUES (7, "Revati", 40, "Female"), (8, "Raghu", 35, "Male"), (9, "Rahul", 25, "Male");
INSERT INTO users VALUES (10, "Rohit", 34, "Male"),(11, "Ramana", 32, "Male"), (12, "Rehman", 55, "Male");
```

```sql
DESC DETAIL users
```

```sql
SELECT * FROM users
```

```sql
SELECT * FROM users WHERE gender = 'Female'
```

```sql
CREATE TABLE users_part (
  id INT,
  name STRING,
  age INT,
  gender STRING
)
PARTITIONED BY (gender)
```

```sql
INSERT INTO users_part
SELECT * FROM users
```

```sql
SELECT * FROM users_part
```

```sql
DESC EXTENDED users_part
```

```sql
SELECT * FROM users_part WHERE gender = 'Female'
```

#### Partitioning Example 2

```sql
%python
dataPath = "dbfs:/FileStore/data/shopping/invoices_201_99457.parquet"

df_invoices = spark.read.parquet(dataPath).repartition(20)

df_invoices.write.mode("overwrite").saveAsTable("default.invoices");
df_invoices.write.mode("overwrite").partitionBy("category").saveAsTable("default.invoices_part");
```

```sql
SHOW TABLES
```

```sql
SELECT * FROM invoices WHERE category = 'Clothing' AND gender = 'Female'
```

```sql
SELECT * FROM invoices_part where category = 'Clothing' and gender = 'Female'
```

```sql
DESC EXTENDED invoices_part
```

```sql
%fs ls dbfs:/user/hive/warehouse/invoices_part
```

```sql
%fs ls dbfs:/user/hive/warehouse/invoices_part/category=Clothing/
```
